## check the outliers

```python
num_col= [i for i in df.columns if df[i].dtypes!='O']
num_col
```

```python
for i in num_col:                               #### Showing the outliers of every columns
    df.boxplot(column=num_col)                  ## Boxplot is must need to find outliers
plt.show()
```

```python
Q1=df.quantile(0.25)
Q1                               ### Steps for removing outliers, Here we can use formulla of  IQR
```

```python
Q3=df.quantile(0.75)
Q3
```

```python
### IQR formullaa used [IQR=Q3-Q1]
IQR=IQR=Q3-Q1
IQR
```

```python
Lower_Extreme= Q1-1.5*IQR                 ## Find the Lower extreme
Lower_Extreme
```

```python
Upper_Extreme= Q1-1.5*IQR                 ## Find the Upper extreme
Upper_Extreme
```

# Step 3: Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.hist(df,bins=11)
plt.show()
```

```python
plt.boxplot(df)
plt.show()
```

```python
## Bar chart analysis for categorical column.#Identify categroical column in the dataset
cat_col= df.select_dtypes(include='object').columns.tolist()
cat_col
```

```python
# Scatter plots or correlation heatmaps to explore relationships between pairs of variables
plt.figure(figsize=(10, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```
