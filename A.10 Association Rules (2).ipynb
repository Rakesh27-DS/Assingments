```python
from mlxtend.frequent_patterns import association_rules, apriori
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_excel('Online retail.xlsx')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
data=pd.get_dummies(df,dtype='int')
data.head()
```

```python
#Convert the data into a list of transactions
transactions = [transaction.split(',') for transaction in data]
print(transactions[:5])
```

```python
te = TransactionEncoder()
```

```python
te_ary = te.fit(transactions).transform(transactions)
```

```python
basket = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
basket.head()
```

```python
freq_items=apriori(basket,min_support=0.01,use_colnames=True)
freq_items
```

```python
association_rules=association_rules(freq_items, metric='lift', min_threshold=5)
```

# Interview questions

# What is lift? Why is it important in Association rules?
Lift indicates the strength of a rule over the random co-occurrence of X & Y

It compares the observed frequency of the items occuring together to the frequency taht would be expected if they were independent

Lift=Number of item X & Y occured /(Total occurence of X)*(Toatal occurence of Y)

*It helps to filter out association rules that are simply due to chance

*It focuses on the rules that reveal genuine relationships between items rather than just common occurences

# What is support and Confidence? How do you caluclate them?

(a)Support is a measure of the number of times an item set appears in a dataset

Support provides fraction of transactions which contains the item X and Y

Support= (number of items X and Y occured)/(Total number of trasactions)

(b)Confidence is measure of likelihood that an itemset will appear if another itemset appears

It indicates how often the items X and Y occur together given the number of times X occur

Confidence= (Number of times X and Y occur)/(Total occurence of X)
